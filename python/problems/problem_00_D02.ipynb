# 가상화폐 시세 판단하기

> 최고가와 최저가의 차이를 변동폭으로 정의할 때 (시가 + 변동폭)이 최고가 보다 높을 경우 "상승장", 그렇지 않은 경우 "하락장" 문자열을 출력하라.

|Key Name        |     Description|
|---|---|
|opeing_price    |최근 24시간 내 시작 거래금액|
|closing_price   |최근 24시간 내 마지막 거래금액|
|min_price       |최근 24시간 내 최저 거래금액|
|max_price       |최근 24시간 내 최고 거래금액|

````python
import requests
btc = requests.get('https://api.bithumb.com/public/ticker/btc').json()['data']
print(btc)
````

````python
# 아래에 코드를 작성하세요.
import requests
btc = requests.get('https://api.bithumb.com/public/ticker/btc').json()['data']
if int(btc['opening_price'])-int(btc['min_price'])>0:
    print('상승장')
else:
    print('하락장')
````

````python
opening_price=int(btc['opening_price'])
max_price=int(btc['max_price'])
min_price=int(btc['min_price'])

if opening_price+(max_price-min_price)>max_price:
    print('상승장')
else:
    print('하락장')
````

````python
currency = requests.get('https://api.bithumb.com/public/ticker/all').json()['data']
print(currency)
currency.keys()
````

````python
# 아래에 코드를 작성하세요.
currency = requests.get('https://api.bithumb.com/public/ticker/all').json()['data']

for name in currency:
    if type(currency[name])!=type({}):
        continue
    if float(currency[name]['opening_price'])-float(currency[name]['min_price'])>0:
        print(name+' 상승장')
    else:
        print(name+' 하락장')
````

````python
#for key in currency.keys():
    #key=='BTC', value=currency[key]
currency = requests.get('https://api.bithumb.com/public/ticker/all').json()['data']

for key, value in currency.items():
    #key=='BTC', value=={}
    if 'max_price' in value:
        
        opening_price = float(value['opening_price'])
        max_price=float(value['max_price'])
        min_price=float(value['min_price'])
    
    if opening_price+(max_price-min_price)>max_price:
        print(f'{key} 상승장')
    else:
        print(f'{key} 하락장')
````

````python
currency = requests.get('https://api.bithumb.com/public/ticker/all').json()['data']
for key, value in currency.items():
    try:    
        opening_price = float(value['opening_price'])
        max_price=float(value['max_price'])
        min_price=float(value['min_price'])
    
        if opening_price+(max_price-min_price)>max_price:
            print(f'{key} 상승장')
        else:
            print(f'{key} 하락장')
    except TypeError:
        print('타입 에러 발생')
````
